Forecasting using fbprophet algo

```python
#date dd/mm/yyyy to yyyy-mm-dd
date = pd.to_datetime(air['Date'], format='%d/%m/%Y').dt.strftime('%Y-%m-%d')
print(date)
```

```python
time=air['Time']
print(time)
```

```python
time=time.apply(lambda x: x.replace('.',':'))
print(time)
```

```python
#combining 2 series to a dataframe
date_time=pd.concat([date,time],axis=1)
print(date_time)
```

```python
#combining date and time
date_time['ds']=date_time['Date'].astype(str)+' '+date_time['Time'].astype(str)
date_time.head()
```

```python
#convert ds to date time format from object datatype
data=pd.DataFrame()
```

```python
data['ds']=pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

```python
data.info()
```

```python
data['y']=air['RH'] #RELATIVE HUMIDITY
```

```python
data.head()
```

```python
! pip install prophet
```

```python
from prophet import Prophet
```

```python
# training prophet model
model=Prophet()
model.fit(data)
```

```python
future=model.make_future_dataframe(periods=365,freq='H')
future.tail()
```

```python
forecast=model.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
fig=model.plot(forecast)
```

```python
fig1=model.plot_components(forecast)
```
